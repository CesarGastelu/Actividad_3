```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Ventas_totales_sin_nulos.csv')
df = data.copy()
```

```python
df.columns
```

```python
fig = plt.figure(figsize=(10,6))
plt.boxplot(df['ventas_precios_constantes'])
plt.title('Outliers de ventas precios constantes')
plt.show()
```

```python
y = df['ventas_precios_constantes']

percentile25 = y.quantile(0.25)
percentile75 = y.quantile(0.75)
iqr = percentile75 - percentile25

limiteSuperiorIQR = percentile75 + 1.5*iqr
limiteInferiorIQR = percentile25 - 1.5*iqr

dataClean_VentasPreciosConstantes = df[(y <= limiteSuperiorIQR) & (y >= limiteInferiorIQR)]

fig = plt.figure(figsize=(10,6))
plt.boxplot(dataClean_VentasPreciosConstantes['ventas_precios_constantes'])
plt.title('Outliers de ventas precios constantes')
plt.show()
```
